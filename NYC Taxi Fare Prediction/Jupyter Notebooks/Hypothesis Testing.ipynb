```python
nyc_taxi['payment_type'] = nyc_taxi['payment_type'].astype('category')
```

```python
nyc_taxi.groupby('payment_type', observed=False)['fare_amount'].mean()
```

Based on the averages, it appears that customers who pay with credit card tend to pay a larger fare amount than customers who pay via cash. However, this difference might arise from random sampling, rather than being a true difference in fare amount. To assess whether the difference is statistically significant, let's conduct a hypothesis test.

$H_0$: There is no difference in the average fare amount between customers who use credit cards and customers who use cash.

$H_A$: There is a significant difference in the average fare amount between customers who use credit cards and customers who use cash.

Let's assume a significance level of 5%.

```python
credit_card_trips = nyc_taxi[nyc_taxi['payment_type'] == 1]['fare_amount']
cash_trips = nyc_taxi[nyc_taxi['payment_type'] == 2]['fare_amount']
t_stat, pval = stats.ttest_ind(a=credit_card_trips, b=cash_trips, equal_var=False)
pval
```

The p-value is significantly smaller than the pre-decided significance level of 5%.
Therefore, we'll reject the null hypothesis. 

*Thus, we conclude that there is a statistically significant difference in the average fare amount between customers who use credit cards and customers who use cash.*

**Key Insight:**
Encouraging customers to pay with credit cards could generate more revenue.
